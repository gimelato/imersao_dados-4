```python
#Desafio 4 - Pegar outras estatísticas dos dados (média, mediana, min e max) e analisar a saída

dados.describe()

#Gosto de comparar a média(mean) com a mediana(50%) para ter noção se a distribuição dos dados é uniforme.
# Valores distantes de média e mediana indicam não normalidade na distribuição, ou seja, neste caso
# provavelmente há alguns poucos valores muito altos de metragem jogando a média para cima

# Valor máximo de 40 banheiros e 50 vagas??? 750000 metros???? Qual o tipo desses imóveis? Eles deveriam ser considerados nas análises?
```

```python
# Desafio 2 - Procurar outras duas formas de selecionar os dados por bairro

#Primeiramente aprendendo os usos do loc e iloc
dados.iloc[:, 1]
```

```python
dados.loc[:,"Bairro"]
```

```python
#Aplicando o loc para selecionar apenas os imóveis da Vila Mariana
dados.loc[lambda dados: dados['Bairro'] == "Vila Mariana", :]
```

```python
#Utilizando o loc apenas numa série (neste caso, a coluna de bairros do dataframe)
dados['Bairro'].loc[lambda s: s == 'Vila Mariana']
```

```python
#Utilizando o método query

dados.query('Bairro == "Vila Mariana" or Bairro == "Mooca"')

#Aqui eu quis adicionar um bairro para adicionar um operador na função
```

```python
#Desafio 1- Calcular a média da metragem dos imóveis por bairro

dados.groupby("Bairro")["Metragem"].mean()

# a função groupby agrupa os dados pelo(s) parâmetros dados. Então selecionei a metragem e calculei a média
```

```python
# Desafio 3- Explorar alguns gráficos na documentação e aplicar nas demais colunas do DF e tentar colocar alguma conclusão da análise feita

#Como eu fiquei muito intrigada com os 40 banheiros, vou brincar com gráficos que me mostrem os valores de banheiros
dados["Banheiros"].plot.hist(bins=40, figsize=(7, 5));

#Olhando para o gráfico pareceu ser um valor pontual e totalmente fora da curva (ainda que 14 banheiros ainda seja muita coisa)
```

```python
#Quis fazer um scatterplot para ver se tem alguma relação entre o número de banheiros e vagas
dados.plot.scatter(x="Banheiros", y="Vagas");

#Parecem que os valores baixos de banheiro se relacionam com os valores baixos de vagas,
# mas se tratando de imóveis que imagino em sua maioria serem residenciais, são de fato poucos banheiros e poucas vagas.
# Como o local com 40 banheiros não tem muitas vagas imagino que seja um erro de registro.
# Tem muitos locais com muitas vagas e menor quantia de banheiros. Onde tem 50 vagas e 10 banheiros?
```

```python
#Adicionei gradiente de cor conforme o número de quartos
dados.plot.scatter(x="Banheiros", y="Vagas", c="Quartos", colormap="gist_rainbow", figsize=(10, 6));

#Os locais com maior número de quartos não tem o maior número de vagas nem de banheiros
```
